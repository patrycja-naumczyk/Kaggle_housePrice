### Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Load the data

```python
df=pd.read_csv("train.csv")
df.head()
```

```python
df.shape
```

### Inspect the data


#### Data types (remove object types)

```python
pd.set_option('display.max_rows', 90)
print(df.dtypes)
```

```python
for col in df.columns:
    if df[col].dtype == "object":
        df[col] = pd.Series(df[col], dtype="string")
```

```python
print(df.dtypes)
```

```python
df[''] = pandas.Series(df['id'], dtype="string")
```

#### Drop columns with missing values

```python
df.isnull().sum()
```

```python
for col in df.columns:
    if df[col].isnull().sum()>0:
        df.drop(col, axis=1, inplace=True)
```

```python
df.shape
```

#### Exploratory visualizations

##### Categorical variables

```python
for col in df.columns:
    if df[col].dtype == "string":
        sns.catplot(x="SalePrice", y=col, data=df)
        plt.ylabel(col,fontsize=20)
        plt.xlabel("Sale Price",fontsize=20)
        plt.locator_params(axis='x', nbins=7)
        plt.show()
```

```python
df.drop(["Street", "LotShape", "LandContour", "Utilities", "LotConfig", "LandSlope", "Condition1", "Condition2", "RoofStyle", "RoofMatl", "ExterCond", "Heating", "Functional", "SaleType"], axis=1, inplace=True)
```

```python
df.shape
```

```python
for col in df.columns:
    if df[col].dtype == "string":
        g = sns.catplot(x="SalePrice", y=col, data=df, kind="violin")
        sns.swarmplot(x="SalePrice", y=col, color="k", size=1, data=df, ax=g.ax)
        plt.ylabel(col,fontsize=20)
        plt.xlabel("Sale Price",fontsize=20)
        plt.locator_params(axis='x', nbins=7)
        plt.show()
```

```python
for col in df.columns:
    if df[col].dtype == "int64" or df[col].dtype == "float64":
        sns.relplot(y="SalePrice", x=col, data=df, kind="scatter")
        plt.xlabel(col,fontsize=20)
        plt.ylabel("Sale Price",fontsize=20)
        plt.locator_params(axis='x', nbins=7)
        plt.show()
```
